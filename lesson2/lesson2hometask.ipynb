```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(2023)
values = np.random.randint(1, 100, 20)

def deepest_lake(values):  
    values_for_use = values.copy()
    
    first_index = np.argmax(values_for_use)
    values_for_use[first_index] = -1
    
    second_index = np.argmax(values_for_use)
    
    return sorted([first_index, second_index])

indices = deepest_lake(values)

number_between_highest_points = values[indices[0]:indices[1] + 1]

```

```python
plt.figure(figsize=(10, 5))
plt.plot(values, color='b', marker='o', label=" ")
plt.plot(range(indices[0], indices[1] + 1), number_between_highest_points, color='red', marker='o', label="  ")
plt.xlabel("")
plt.ylabel("")
plt.title("     ")
plt.legend()
plt.show()
```
