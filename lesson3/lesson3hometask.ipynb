```python
import pandas as pd
import numpy as np

def answer_one():
    data = pd.read_excel("Energy Indicators.xls", skiprows = 17, usecols = "C,D,E,F", skipfooter = 38, names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'])
    energy = pd.DataFrame(data)
    energy.replace("...", np.nan, inplace=True)
    energy['Energy Supply'] = energy['Energy Supply']*1000000
    energy['Country'] = energy['Country'].str.replace(r'\d{1,2}$', '', regex = True).str.strip()
    energy['Country'] = energy['Country'].str.replace(r'\(.*\)', '', regex = True).str.strip()
    value_for_replacing = ["Republic of Korea", "United States of America", "United Kingdom of Great Britain and Northern Ireland", 'China, Hong Kong Special Administrative Region']
    required_values = ["South Korea", "United States", "United Kingdom", "Hong Kong"]
    energy.replace(value_for_replacing, required_values, inplace=True)
    gdp_data = pd.read_csv("world_bank.csv", sep=',', skiprows=4)
    GDP = pd.DataFrame(gdp_data)
    value_to_replace = ['Korea, Rep.','Iran, Islamic Rep.','Hong Kong SAR, China']
    replacement_value = ['South Korea','Iran','Hong Kong']
    GDP.replace(value_to_replace, replacement_value, inplace = True)
    ScimEn_data = pd.read_excel("scimagojr country rank 1996-2023.xlsx")
    ScimEn = pd.DataFrame(ScimEn_data)
    df = pd.merge(energy,ScimEn,how = 'inner', left_on='Country', right_on='Country')
    df = pd.merge(df,GDP,how = 'inner', left_on='Country', right_on='Country Name')
    df = df[df['Rank']<=15]
    df = df[['Country','Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    return df

answer_one()
```

```python
def answer_two():
    Top15 = answer_one()
    Top15.sort_values("Rank", inplace=True)
    list = Top15.iloc[:, 11:22]
    list_avg = [0for _ in range(15)]
    for i in range(len(list.values.tolist())):
        for j in list.values.tolist()[i]:
            list_avg[i]+=j
        list_avg[i]/=10
    list = Top15["Country"].to_list()
    avgGDP = pd.Series(data=list_avg, index=list).sort_values(ascending=False)
    return avgGDP
answer_two()    
```

```python
def answer_three():
    country_name = answer_two().index[6]
    country_data = answer_one().set_index("Country").loc[country_name]
    return int(country_data["2006"]-country_data["2015"])
answer_three()
```

```python
def answer_four():
    Top15 = pd.DataFrame()
    Top15 = answer_one()[['Country', 'Citations', 'Self-citations']].copy()
    Top15.insert(3, "Ratio", np.nan)
    Top15['Ratio']=Top15['Self-citations']/Top15['Citations']
    Top15 = Top15.sort_values(ascending=False, by='Ratio')
    tuple = (Top15['Country'].iloc[0], float(Top15['Ratio'].iloc[0]))
    return tuple
answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['Estimation'] = Top15['Energy Supply']/Top15['Energy Supply per Capita']
    Top15 = Top15.sort_values(by='Estimation', ascending=False)
    result = Top15.iloc[2]['Country']
    return result
answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable per capita'] = Top15['Citable documents'] / Top15['Population']
    return float(Top15['Citable per capita'].corr(Top15['Energy Supply per Capita'], method='pearson'))
answer_six()
```

```python
import pandas as pd

def answer_seven():
    Top15 = answer_one()  # Переконайтеся, що ця функція повертає DataFrame зі стовпцем 'Country'
    
    ContinentDict = {
        'China': 'Asia', 
        'United States': 'North America', 
        'Japan': 'Asia', 
        'United Kingdom': 'Europe', 
        'Russian Federation': 'Europe', 
        'Canada': 'North America', 
        'Germany': 'Europe', 
        'India': 'Asia',
        'France': 'Europe', 
        'South Korea': 'Asia', 
        'Italy': 'Europe', 
        'Spain': 'Europe', 
        'Iran': 'Asia',
        'Australia': 'Australia', 
        'Brazil': 'South America'
    }

    if 'Country' in Top15.columns:
        Top15.set_index('Country', inplace=True)
    Top15['Continent'] = Top15.index.map(ContinentDict)
    Top15.dropna(subset=['Continent'], inplace=True)
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    result = Top15.groupby('Continent')['Population'].agg(['size', 'sum', 'mean', 'std'])
    return result

answer_seven()
```
